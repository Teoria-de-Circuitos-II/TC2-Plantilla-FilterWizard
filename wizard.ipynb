# Ejemplo de uso de plot_helper.py

Bode y respuesta de un seno de la transferencia:

$$H(s) = \frac{5(s - 1)(s - 2)}{(s + 3)(s + 4)}$$

```python
from plot_helper import create_figure, plot_time, plot_pzmap, plot_bode, plot_plantilla

tf = ss.lti([1, 2], [-3, -4], 5)

w, mag, ph = ss.bode(tf, n=1000)

fig, ax = create_figure(10, 5, axes_num=2)

ax_mag = ax[0]
ax_ph = ax[1]

plot_bode(ax_mag, ax_ph, mag, ph, w/ (2 * np.pi))
plt.show()
```

```python
w1 = 2*2*np.pi
w2 = 1.5*2*np.pi

t = np.linspace(0, 4*2 * np.pi, 10000)
y1 = np.sin(w1 * t)
y2 = np.sin(w2 * t)

_, y1, _ = ss.lsim(tf, y1, t)
_, y2, _= ss.lsim(tf, y2, t)
fig, ax = create_figure(10, 5)
plot_time(ax, t, y1, y2, label1='Seno de 2Hz', label2='Seno de 1,5Hz')
plt.show()
```
